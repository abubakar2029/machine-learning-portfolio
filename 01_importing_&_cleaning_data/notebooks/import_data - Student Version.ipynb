```python
# Preparing a sample SQL db
import sqlite3
import pandas as pd

# create a database file
conn = sqlite3.connect("test.db")

# sample dataframe
df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Ali", "Sara", "John"],
    "score": [90, 85, 78]
})

# write dataframe to SQL table
df.to_sql("students", conn, if_exists="replace", index=False)
conn.close()
```
